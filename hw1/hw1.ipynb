4.資料分割與建置迴歸模型

```python
from sklearn.model_selection import train_test_split
X=df.loc[:,'bedrooms':'sqft_lot15']
y=df['Price']
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=100,train_size=0.8,test_size=0.2)
df_count=pd.concat([y.value_counts(),y_train.value_counts(),y_test.value_counts()],axis=1)
df_count.columns=['y','y_train','y_test']
df_count.head()     
```

```python
#線性回歸
from sklearn.linear_model import LinearRegression

x,y=df.loc[:,['sqft_living']],df.loc[:,['Price']]
lr=LinearRegression()
lr.fit(x,y)
print('linearModel.coef_=',lr.coef_)
print('linearModel.intercept_ =',lr.intercept_)
sns.set()
sns.set_palette(sns.color_palette(flatui))
plt.scatter(x,y,facecolor='xkcd:azure',edgecolor='black',s=20)
plt.xlabel('sqft_living')
plt.ylabel('Price')

n_x=np.linspace(x.min(),x.max(),100)
n_y=lr.intercept_+lr.coef_[0]*n_x
plt.plot(n_x,n_y,color='r',lw=3)
```

5.模型效能評估

```python
#MSE
from sklearn.metrics import mean_squared_error

lr=LinearRegression()
lr.fit(X_train,y_train)
y_train_pred=lr.predict(X_train)
y_test_pred=lr.predict(X_test)
print('MSE(training): %.3f, MSE(testing): %.3f' %(
    mean_squared_error(y_train,y_train_pred),
    mean_squared_error(y_test,y_test_pred)))
```

```python
#決定係數R^2
from sklearn.metrics import r2_score
print('R^2(training): %.3f,R^2(testing): %.3f' %(
    r2_score(y_train,y_train_pred),
    r2_score(y_test,y_test_pred)))
    
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
print('MAE(training): %.3f, MAE(testing): %.3f' %(
    mean_absolute_error(y_train,y_train_pred),
    mean_absolute_error(y_test,y_test_pred)))
```

```python
#Shapiro-Wilk常態性檢定
import statsmodels.api as sm
from scipy.stats import shapiro
x,y=df.loc[:,'sqft_living':'sqft_lot15'],df.loc[:,['Price']]
x=sm.add_constant(x)
model=sm.OLS(y,x).fit()
stat,p=shapiro(model.resid)
print('Statistics: %.3f,p-value: %.3f' %(stat,p))
alpha=0.05
if p>alpha:
    print('看起來是常態分布(無法拒絕H0)')
else:
    print('看起來不是常態分布(拒絕H0)')
```

```python
#Kolmogorov-Smirnov常態性檢定
from scipy.stats import kstest
stat,p=kstest(model.resid,'norm')
print('Statistics: %.3f,p-value: %.3f'%(stat,p))
alpha=0.05
if p>alpha:
    print('看起來是常態分布(無法拒絕H0)')
else:
    print('看起來不是常態分布(拒絕H0)')
```

```python
#QQ圖
from statsmodels.graphics.gofplots import qqplot
sns.set()
sns.set_palette(sns.color_palette(flatui))
plt.figure(figsize=(15, 10), dpi=40)
qqplot(model.resid,line='r');
plt.title("QQ plot", fontsize = 20, fontweight = 'bold')
plt.show()
```

```python
#檢測獨立性-Durbin-Watson
from statsmodels.stats.stattools import durbin_watson

dw=durbin_watson(model.resid)
print('dw: %.3f' %dw)

if 2<= dw <=4:
    print('誤差項獨立')
elif 0 <= dw <2:
    print('誤差項不獨立')
else:
    print('計算錯誤')
```

```python
#殘差圖
from sklearn.preprocessing import StandardScaler
import seaborn as sns
sns.set()
sns.set_palette(sns.color_palette(flatui))
df_resid=pd.DataFrame()
df_resid['y_pred']=model.predict(x)
df_resid['resid']=model.resid
df_resid=StandardScaler().fit_transform(df_resid)

kws={'color':'red','lw':3}
sns.residplot(x=df_resid[:,0],y=df_resid[:,1],lowess=True,line_kws=kws)
plt.title('The residual plot',fontsize=14)
plt.xlabel('Predicted Values(Standardization)',fontsize=14)
plt.ylabel('Residual(Standardization)',fontsize=14)
plt.show()
```

7.非線性迴歸模型 or 特徵變數變換

```python
#二階多項式迴歸
from sklearn.preprocessing import PolynomialFeatures

X=df.loc[:, 'bedrooms': 'sqft_lot15'].values
y=df['Price'].values
#將特徵轉成多項式
pol_d=PolynomialFeatures(degree=2)
X_poly=pol_d.fit_transform(X)
#訓練集和測試集
X_train, X_test, y_train, y_test = train_test_split(
    X_poly, y, test_size=0.2, random_state=0)
#建立線性迴歸模型，並在訓練集上擬合
lr=LinearRegression()
lr.fit(X_train, y_train)
#在訓練集和測試集上進行預測
y_train_pred=lr.predict(X_train)
y_test_pred=lr.predict(X_test)
#計算r2 score
print('R^2(training): %.3f,R^2(testing): %.3f' %(
    r2_score(y_train,y_train_pred),
    r2_score(y_test,y_test_pred)))
#計算MSE
print('MSE(training): %.3f, MSE(testing): %.3f' %(
    mean_squared_error(y_train,y_train_pred),
    mean_squared_error(y_test,y_test_pred)))
#計算MAE
print('MAE(training): %.3f, MAE(testing): %.3f' %(
    mean_absolute_error(y_train,y_train_pred),
    mean_absolute_error(y_test,y_test_pred)))
#計算訓練和測試資料的預測殘差
y_train_resid=y_train_pred - y_train
y_test_resid=y_test_pred - y_test 

sns.residplot(y_train_pred, y_train_resid, lowess=True, 
              color="skyblue", label='Training data', 
              scatter_kws={'s': 25, 'alpha':0.7}, 
              line_kws={'color': 'b', 'lw':2})
sns.residplot(y_test_pred, y_test_resid, lowess=True, 
              color="yellowgreen", label='Testing data', 
              scatter_kws={'s': 25, 'marker':'x'}, 
              line_kws={'color': 'g', 'lw':2})
plt.legend()
plt.xlabel('Predicted Price')
plt.ylabel('Residual')
plt.show()
```

```python
#隨機森林
from sklearn.ensemble import RandomForestRegressor

X=df.loc[:,'bedrooms':'sqft_lot15'].values
y=df['Price'].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0)

reg=RandomForestRegressor(n_estimators=100,# default 
                            criterion='mse', # default
                            random_state=0, 
                            n_jobs=-1)
reg.fit(X, y)
y_train_pred=reg.predict(X_train)
y_test_pred=reg.predict(X_test)
#計算R^2
print('R^2(training): %.3f,R^2(testing): %.3f' %(
    r2_score(y_train,y_train_pred),
    r2_score(y_test,y_test_pred)))

#計算MSE
print('MSE(training): %.3f, MSE(testing): %.3f' %(
    mean_squared_error(y_train,y_train_pred),
    mean_squared_error(y_test,y_test_pred)))
#計算MAE
print('MAE(training): %.3f, MAE(testing): %.3f' %(
    mean_absolute_error(y_train,y_train_pred),
    mean_absolute_error(y_test,y_test_pred)))

y_train_resid=y_train_pred - y_train
y_test_resid=y_test_pred - y_test 

sns.residplot(y_train_pred, y_train_resid, lowess=True, 
              color="skyblue", label='Training data', 
              scatter_kws={'s': 25, 'alpha':0.7}, 
              line_kws={'color': 'b', 'lw':2})
sns.residplot(y_test_pred, y_test_resid, lowess=True, 
              color="yellowgreen", label='Testing data', 
              scatter_kws={'s': 25, 'marker':'x'}, 
              line_kws={'color': 'g', 'lw':2})
plt.legend()
plt.xlabel('Predicted Price')
plt.ylabel('Residual')
plt.show()
```

8.迴歸模型之正規化

```python
#脊迴歸
from sklearn.preprocessing import StandardScaler
#標準化
scalar = StandardScaler()
X = scalar.fit_transform(df.iloc[:, :-1])
y = scalar.fit_transform(df.loc[:, ['Price']])
X[:2, :]
```

```python
from sklearn.linear_model import Ridge, RidgeCV

alphas=np.logspace(-3, 3, 50)
reg_cv=RidgeCV(alphas, store_cv_values=True)
reg_cv.fit(X,y)
print('Best alpha: %.3f'% reg_cv.alpha_)
```

```python
scores = np.mean(reg_cv.cv_values_, axis=0)[0]
scores_std = np.std(reg_cv.cv_values_, axis=0)[0]

# 設定 x 軸為對數
plt.semilogx(alphas, scores, color='red')

# 繪製 MSE +/- 標準誤(std error)
std_error = scores_std / np.sqrt(len(scores))
plt.semilogx(alphas, scores+std_error, 'b--')
plt.semilogx(alphas, scores-std_error, 'b--')
plt.fill_between(alphas, scores+std_error, 
                 scores-std_error, alpha=0.2)
plt.title('Ridge Regression',fontsize=20)
plt.ylabel('CV MSE +/- std error')
plt.xlabel(r'$\alpha$')
plt.xlim([alphas[0], alphas[-1]])
```

```python
reg = Ridge(alpha=reg_cv.alpha_)
y_pred = reg.fit(X, y).predict(X)
print('MSE =', mean_squared_error(y, y_pred))
print('R^2:',r2_score(y,y_pred))
print('MAE:',mean_absolute_error(y,y_pred))
      
```

```python
# 因為已經標準化，所以把截距拿掉
model = Ridge(alpha=reg_cv.alpha_,
              fit_intercept=False)
model.fit(X, y)
df_coef = pd.DataFrame(data=model.coef_, 
                       columns=df.columns[:-1])
df_coef
```

```python
coef = np.array(df_coef)[0]
# 依迴歸係數的絕對值排序(由大到小)
coef_abs = np.apply_along_axis(abs, 0, coef)
idx = np.argsort(-coef_abs)
sns.set()
sns.set_palette(sns.color_palette(flatui))
plt.bar(range(len(coef)), coef[idx])
plt.xticks(range(len(coef)), df_coef.columns[idx], 
           rotation=60)
plt.ylabel('weights');
```

```python
#套索迴歸
from sklearn.linear_model import LassoCV, Lasso

alphas = np.logspace(-5, 3, 50)
reg_cv = LassoCV(alphas=alphas, cv=10, n_jobs=-1)
reg_cv.fit(X, y.reshape(-1))
print('Best alpha: %.6f' % reg_cv.alpha_)
```

```python
scores = np.mean(reg_cv.mse_path_, axis=1)
scores_std = np.std(reg_cv.mse_path_, axis=1)

# 設定 x 軸為對數
plt.semilogx(alphas, scores, color='red')

# 繪製 MSE +/- 標準誤(std error)
std_error = scores_std / np.sqrt(len(scores))
plt.semilogx(alphas, scores+std_error, 'b--')
plt.semilogx(alphas, scores-std_error, 'b--')
plt.fill_between(alphas, scores+std_error, 
                 scores-std_error, alpha=0.2)
plt.title('Lasso Regression',fontsize=20)
plt.ylabel('CV MSE +/- std error')
plt.xlabel(r'$\alpha$')
plt.xlim([alphas[0], alphas[-1]]);
```

```python
reg = Lasso(alpha=reg_cv.alpha_,fit_intercept=False)
y_pred = reg.fit(X, y).predict(X)
print('MSE =', mean_squared_error(y, y_pred))
print('R^2:',r2_score(y,y_pred))
print('MAE:',mean_absolute_error(y,y_pred))
      
```

```python
model = Lasso(alpha=reg_cv.alpha_,
              fit_intercept=False)
model.fit(X, y)
df_coef = pd.DataFrame(data=model.coef_.reshape(1,-1), 
                       columns=df.columns[:-1])
df_coef
```

```python
coef = np.array(df_coef)[0]
# 依迴歸係數的絕對值排序(由大到小)
coef_abs = np.apply_along_axis(abs, 0, coef)
idx = np.argsort(-coef_abs)
sns.set()
sns.set_palette(sns.color_palette(flatui))
plt.bar(range(len(coef)), coef[idx])
plt.xticks(range(len(coef)), df_coef.columns[idx], 
           rotation=60)
plt.ylabel('weights');
```

```python
#彈性網
from sklearn.linear_model import ElasticNetCV,ElasticNet
# 僅設定 rho 的範圍，alpha 的範圍採自動設定
rho = [.1, .5, .7, .9, .95, .99, 1]
reg_cv = ElasticNetCV(cv=10, l1_ratio=rho)
reg_cv.fit(X, y.reshape(-1))
sns.set()
sns.set_palette(sns.color_palette(flatui))
df_coef = pd.DataFrame(data=reg_cv.coef_.reshape(1,-1), 
                       columns=df.columns[:-1])
coef = np.array(df_coef)[0]
# 依迴歸係數的絕對值排序(由大到小)
coef_abs = np.apply_along_axis(abs, 0, coef)
idx = np.argsort(-coef_abs)
plt.title('Elastic Net',fontsize=20)
plt.bar(range(len(coef)), coef[idx])
plt.xticks(range(len(coef)), df_coef.columns[idx], 
           rotation=60)
plt.ylabel('weights');
```
